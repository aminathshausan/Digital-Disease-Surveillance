```python
import pandas as pd
from datetime import date
from pytrends.request import TrendReq
```

```python

#define list of keywords and states
kw_lst = ["coronavirus", "covid",   #covid terms
           "influenza", "Haemophilus influenzae",  #influenza terms
           "flu", "parainfluenza", "H1N1", "H7N9", "H5N1", "H3N2", 
           "grippe", "gripe", "adenovirus", "rhinovirus", 
           "respiratory syncytial virus", "metapneumovirus", 
          "Bordetella pertussis", "Mycoplasma pneumoniae", "pneumonia",
           "bronchitis", "H9N2", "sinusitis", 
           "upper respiratory tract infection", "Tamiflu"]

state_lst = ["AU", "AU-NSW", "AU-QLD","AU-SA",
         "AU-TAS", "AU-VIC", "AU-WA", "AU-ACT", "AU-NT"]
```

```python
pytrends = TrendReq(hl='en-US', tz=360)

#use this if you are blocked by rate limit
# pytrends = TrendReq(hl='en-US', tz=360, timeout=(10,25), 
#                     proxies=['https://35.201.123.31:880',], 
#                     retries=2, backoff_factor=0.1, 
#                     requests_args={'verify':False})

#build a function to scrap data
def get_data(key_word, state):
  
    pytrends.build_payload([key_word], cat=0, 
                   timeframe='2020-02-01 2023-01-01', 
                           geo=state, gprop='')
    df = pytrends.interest_over_time()
    if not df.empty:
        df.drop('isPartial', axis=1, inplace=True)
        df = df.reset_index()
        
    return df


```
